## 서울시 운동을 하지 않는 이유 통계
### excel 파일을 읽기 - pandas.read_excel(파일)

```python
import pandas as pd
import warnings
warnings.simplefilter('ignore')

# 제목행의 1번 인덱스만 출력(0번은 숨김)
not_exercise = pd.read_excel("./datas/reason_not_exercise.xlsx", header=1)
not_exercise
```

```python
not_exercise.columns[0]
```

```python
not_exercise.drop(columns='시점', inplace=True)
```

```python
not_exercise
```

```python
# 칼럼명 변경 - 객체.rename()
not_exercise.rename(columns={'구분별(1)':'대분류', '구분결(2)':'분류'}, inplace=True)
not_exercise
```

```python
# 행 삭제 - 객체.drop(index=range(시작행, 끝행-1))
not_exercise.drop(index=range(22, 52), inplace=True)
not_exercise
```

```python
# 2행의  NaN를 '성별'로 변경
not_exercise.loc[2, '대분류'] = '성별'
not_exercise
```

```python
not_exercise['대분류'] == '성별'
```

```python
not_ex_sex = not_exercise[not_exercise['대분류'] == '성별'].copy() #깊은 복사
not_ex_sex
```

```python
# 대부분 칼럼 삭제
not_ex_sex.drop(columns='대분류', inplace=True)
not_ex_sex
```

```python
# index를 '분류'로 세팅
not_ex_sex.set_index('분류', inplace=True)
not_ex_sex
```
